```python
def mean_psnr(origin, corrupted, noise):
    psnrs = []
    for i in range(len(noise)):  
        psnr = compare_psnr(origin[i], (corrupted[i] - noise[i]).reshape(512,512))
        psnrs.append(psnr)
    return np.mean(psnrs)
```

```python
def one_one_correspondence( file_list_1,file_list_2 ):
    
    for index1, file1 in enumerate(file_list_1):
        file1 = file1.split('_')
        #print (file1)
        for index2, file2 in enumerate(file_list_2):
            file2 = file2.split('.')
            file2_compare_tag = file2[0].split('_')
            #print (file2)
            if file1[0]==file2_compare_tag[0]:
                temp = file_list_2[index2]
                file_list_2[index2] = file_list_2[index1]
                file_list_2[index1] = temp
    return file_list_1, file_list_2
```

```python
def read_image_in_folder( folder_name_1, folder_name_2 ):
    I1 = []
    I2 = []
    images_from_folder1 = os.listdir( folder_name_1 )
    images_from_folder2 = os.listdir( folder_name_2 )
    [images_from_folder1,images_from_folder2] = one_one_correspondence( images_from_folder1,images_from_folder2 )
    for image in images_from_folder1:
        if( image != ".DS_Store" ):
            I1.append(mpimg.imread(folder_name_1 + image))
    for image in images_from_folder2:
        if( image != ".DS_Store" ):
            I2.append(mpimg.imread(folder_name_2 + image))
    return np.array(I1), np.array(I2)
```

```python
def get_noise( clean_images, distorted_images ):
    assert len(clean_images) == len(distorted_images)
    
    noises = []
    for idx in range(len(clean_images)):
        new = distorted_images[idx] - clean_images[idx]
        #print(n)
        noises.append(new)
    return np.array(noises)
```

```python
def crop_to_patches( dataset, patch_size ):
    # repeatition
    result = []
    step = 10
    for data in dataset: 
        cropped = view_as_windows(data, (patch_size,patch_size), step=step)
        result.append(cropped.reshape(-1, patch_size, patch_size))
    return np.array(result) 
```

```python
def reconstruct( pacthes_divided_by_images ):
    for image in pacthes_divided_by_images:
        elem_num = int(sqrt(len(image)))
        patch_size = image.shape[1]
        final = [[] for _ in range(480)]
        for m in range(elem_num):#each row of an image
            row = image[(m*elem_num):((m+1)*elem_num)]
            for n in row:#each patch
                for idx,r in enumerate(n):
                    for c in r:
                        final[patch_size * m + idx].append(c)
    return final
```

```python
DataSet = namedtuple("DataSet", ["origin", "corrupted", "noise",
                                 "corrupted_patches","noise_patches"])
```

```python
patch_size = 40

natural_dataset = read_image_in_folder( "natural/clean/","natural/distorted/" )
natural_noise = get_noise(natural_dataset[0], natural_dataset[1])
natural_dataset = DataSet(natural_dataset[0], natural_dataset[1], 
                          natural_noise,
                          crop_to_patches( natural_dataset[1], patch_size ),
                          crop_to_patches( natural_noise, patch_size )
                         )

syn_dataset = read_image_in_folder( "synthetic/clean/","synthetic/distorted/" )
syn_noise = get_noise(syn_dataset[0], syn_dataset[1])
syn_dataset = DataSet(syn_dataset[0], syn_dataset[1],
                      syn_noise,
                      crop_to_patches( syn_dataset[1], patch_size ),
                      crop_to_patches( syn_noise, patch_size ),
                      )
```

```python
shape_val = 40

X = natural_dataset.corrupted_patches + syn_dataset.corrupted_patches
Y = natural_dataset.noise_patches + syn_dataset.noise_patches
```

```python
trainX, testX, trainY, testY = train_test_split(X, Y, shuffle=True)
```

```python
trainX = np.reshape(trainX, (trainX.shape[0]*trainX.shape[1], shape_val, shape_val, 1))
testX = np.reshape(testX, (testX.shape[0]*testX.shape[1], shape_val, shape_val, 1))
trainY = np.reshape(trainY, (trainY.shape[0]*trainY.shape[1], shape_val, shape_val, 1))
testY = np.reshape(testY, (testY.shape[0]*testY.shape[1], shape_val, shape_val, 1))
```

```python
def activation( name, input ):
    acti = Activation(name)(input)
    return acti
```

```python
def convolutional( filters, kernel_size, input ):

    conv = Conv2D(filters=filters, kernel_size=(kernel_size, kernel_size),
                      padding="same",
                      kernel_initializer="he_normal",
                      data_format="channels_last")(input)
    return conv
```

```python
def batch_norm( input ):
    norm = BatchNormalization(axis=-1)(input)
    return norm
```

```python
def first_layer( input ):
    
    conv_1 = convolutional( 64, 3, input)
    layer_1 = activation( 'relu', conv_1)
    
    return layer_1
```

```python
def middle_layers( repetition, input ):
    
    for x in range(repetition):
        conv_2 = convolutional( 64, 3 ,input )
        batch = batch_norm( conv_2 )
        input = activation( 'relu', batch ) 
        
    layer_2 = input
    return layer_2
```

```python
def last_layer ( input ):
    layer_3 = convolutional( 1, 3, input)
    return layer_3
```

```python
with tf.device('/gpu:0'):
    config = tf.ConfigProto(
        allow_soft_placement=True, 
        log_device_placement=False
    )
    config.gpu_options.per_process_gpu_memory_fraction=1
    config.gpu_options.allow_growth = True
    
    set_session(tf.Session(config=config))


    input_img = Input(shape=(shape_val, shape_val, 1))
    depth = 17

    layers = last_layer(middle_layers(depth - 2, first_layer(input_img)))

    autoencoder = Model(input_img, layers)
    autoencoder.compile(optimizer='sgd', loss='mean_squared_error')
    
    autoencoder.fit(x=trainX,
                y=trainY,
                epochs=50,
                batch_size=128,
                shuffle=True,
                validation_data=(testX, testY),
                #callbacks=[TensorBoard(log_dir='logs', histogram_freq=0, write_graph=False)]
               )
    
```
